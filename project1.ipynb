```python
df = df.drop("Unnamed: 0", axis = 1)
print(df.head())
```

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("05-Oct","5-10")
df.head()
```

```python
#GENDER_DISTRIBUTION
plt.figure(figsize = (5,5))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
#Ethnicgroup_DISTRIBUTION
plt.figure(figsize = (5,5))
ax = sns.countplot(data = df, x = "EthnicGroup")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
#PARENT_EDUCATION_IMPACT
gb = df.groupby("ParentEduc").agg({"MathScore": 'mean', "ReadingScore":'mean', "WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize= (5,5))
sns.heatmap(gb, annot = True)
plt.title("Parent Education Impact")
plt.show()
```

```python
#PARENT_MARITAL_STATUS
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore": 'mean', "ReadingScore":'mean', "WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize= (5,5))
sns.heatmap(gb1, annot = True)
plt.title("Parent Marital Status")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df,x='WklyStudyHours',y='MathScore')
plt.title('Box Plot Outlier for Study Hours')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df,x='WklyStudyHours',y='WritingScore')
plt.title('Box Plot Outlier for Study Hours')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df,x='WklyStudyHours',y='ReadingScore')
plt.title('Box Plot Outlier for Study Hours')
plt.show()
```
